```python
import geopandas as gpd
import geopandas.tools
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
gdf = gpd.read_file('alerts.csv')
gdf
```

```python
gdf = gdf.drop(columns=['Latitude', 'Longitude'])
gdf
```

```python
locations = geopandas.tools.geocode(gdf['Location Address'])
locations
```

```python
locations['geometry'].unique()
```

```python
combined_gdf = pd.concat([gdf, locations], axis=1)
combined_gdf
```

```python
combined_gdf = gpd.GeoDataFrame(combined_gdf, geometry='geometry')
```

```python
combined_gdf.to_file("geocoded_locations.json", driver="GeoJSON")
```

```python
combined_gdf.dropna("lat")
```

```python
locations_dict = {}
for idx, row in combined_gdf.iterrows():
    location_name = row['Location Address']  # Adjust this to the correct column name for location names
    if pd.notna(row['geometry']) and not row['geometry'].is_empty:
        lat = row['geometry'].y
        lng = row['geometry'].x
    else:
        lat = None
        lng = None
    locations_dict[location_name] = {
        "lat": lat,
        "lng": lng,
        "address": row['address']  # Adjust this to the correct column name for address
    }
```

```python

# Save the dictionary to a JSON file
with open("geocoded_locations.json", "w", encoding="utf-8") as f:
    json.dump(locations_dict, f, indent=2)

print("Geocoded locations saved to geocoded_locations.json")
```
